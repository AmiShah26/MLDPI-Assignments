# **Data Visualization Assignment**

### **In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.**

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```
iris = datasets.load_iris()
x, y = iris.data, iris.target
print("Types of species: ",iris.target_names)
```

**<u>Transform Iris data into 3D</u>**

```
pca = decomposition.PCA(n_components=3)
components = pca.fit_transform(x)
```

**<u>Plot 3D Chart</u>**

```
fig = plt.figure(figsize=(9,9))
ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=45, azim=120)
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
  ax.text3D(components[y == label, 0].mean(), components[y == label, 1].mean() + 1.7,
            components[y == label, 2].mean(), name,
            bbox=dict(alpha=.2, edgecolor='k', facecolor='w'))
y = np.choose(y, [0, 1, 2]).astype(float)
ax.scatter(components[:, 0], components[:, 1], components[:, 2], c = y, edgecolor='k')
plt.show()
```
